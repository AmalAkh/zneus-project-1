```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
normalized_features_df = pd.DataFrame(scaler.fit_transform(df[feature_cols]), columns=feature_cols)
```

```python
normalized_features_df
```

```python
df2 = normalized_features_df.join(df["Class"], how='inner', lsuffix='_caller', rsuffix='_other')
```
